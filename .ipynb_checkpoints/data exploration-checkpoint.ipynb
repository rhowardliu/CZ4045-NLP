```python
import pandas as pd
df = pd.read_json('./reviewSelected100.json', lines=True)
```

```python
df.head()
```

```python
df1 = df.drop(['review_id', 'date'], axis=1)
df1
```

```python
df1.duplicated(subset='user_id', keep=False).value_counts()
```

```python
# top_10_users = df1.groupby('user_id').count().sort_values(by = 'stars', ascending=False)[:10]
flatten = lambda l: [item for sublist in l for item in sublist]

top_10_users = df1.groupby('user_id').size().to_frame(name='count').sort_values(by = 'count', ascending=False)[:10].reset_index()
top_10_users_list = flatten(top_10_users.loc[:,['user_id']].values)

top_10_users_list
```

```python
comments1 = df1.loc[df1['user_id'].isin(top_10_users_list),['user_id','stars', 'text']].sort_values(by='user_id')
comments1
```

```python
df1.loc[:,['stars', 'business_id']].groupby('business_id').mean().sort_values(by='stars')
```
